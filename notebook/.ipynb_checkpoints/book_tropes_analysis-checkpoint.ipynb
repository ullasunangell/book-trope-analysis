# Data analysis of book tropes and reader engagement using Goodreads data.

## Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10,6)
```

## Load Dataset
#### Dataset: Goodreads Books Dataset (Kaggle)

* Many entries in the dataset represent book collections or boxed sets, which artificially inflate page counts and distort analysis.
* These entries were removed to ensure that the dataset represents individual books.
* The dataset size was reduced from: 724 → 700 books

```python
df = pd.read_csv("../data/cleaned_goodreads_books_dataset.csv")
df = df[~df['title'].str.contains('collection|set', case=False, na=False)]
df.head()
```

## Summary Statistics
Average page length: 
* 364 pages
* But the maximum page count (over 10,000 pages) which indicates book series

Engagement varies widely:
* some books have few ratings
* others exceed 9 million ratings

**highly skewed popularity distribution**

```python
df.describe()
```

## Correlation Analysis

* Correlation between page count and rating is: 0.14
* This is extremely weak.

Conclusion:
* Book length has very little influence on reader ratings.
* Readers appear to evaluate books based on story quality rather than length.

```python
df[['pages','rating']].corr()
```

## Most Popular Books
The most rated books include major literary and commercial successes such as:

* *The Hunger Games by Suzanne Collins*
* *To Kill a Mockingbird by Harper Lee*
* *The Great Gatsby by F. Scott Fitzgerald*
* *It Ends With Us by Colleen Hoover*
  
These books combine:
* cultural impact
* strong readership
* long-term popularity.

```python
df.sort_values(by="ratings", ascending=False).head(10)
```

## Tropes

### Trope Detection
* Extracting narrative themes from descriptions.

Keywords were used to identify common narrative themes within book descriptions.

Themes analysed include:

* magic
* romance
* murder
* war
* academy settings

This allows investigation of whether certain narrative elements correlate with reader ratings or engagement.

```python
df = df.dropna(subset=["description"])
```

```python
df["magic"] = df["description"].str.contains("magic", case=False, na=False)
df["romance"] = df["description"].str.contains("love", case=False, na=False)
df["murder"] = df["description"].str.contains("murder", case=False, na=False)
df["war"] = df["description"].str.contains("war", case=False, na=False)
df["academy"] = df["description"].str.contains("academy|school", case=False, na=False)
```

### Trope vs Rating
Books containing **magic themes** show slightly higher ratings:

Magic books:

**3.99**

Non-magic books:

**3.90**

This suggests fantasy elements may correlate with slightly stronger reader reception.

```python
df.groupby("magic")["rating"].mean()
```

```python
df.groupby("magic")["ratings"].mean()
```

## Trope Frequency
#### Most common theme:
**Romance (238 books)**
#### Runner-up:
**War (174)**

* This reflects the strong presence of romance and relationship-driven narratives in the dataset.

```python
trope_counts = df[["magic","romance","murder","war","academy"]].sum()
trope_counts
```

```python
trope_summary = pd.DataFrame({
    "trope": ["magic","romance","murder","war","academy"],
    "avg_rating": [
        df[df["magic"]]["rating"].mean(),
        df[df["romance"]]["rating"].mean(),
        df[df["murder"]]["rating"].mean(),
        df[df["war"]]["rating"].mean(),
        df[df["academy"]]["rating"].mean()
    ]
})

trope_summary
```

```python
trope.summary.plot(x="trope", y="avg_rating", kind="bar")
plt.title("Average Rating by Story Trope")
plt.show()
```

## Visualisations

```python
sns.histplot(df["rating"], bins=20, kde=True)

plt.title("Distribution of Goodreads Ratings")
plt.xlabel("Rating")
plt.ylabel("Number of Books")

plt.show()
```

```python
trope_counts.plot(kind="bar")
plt.title("Frequency of Story Tropes")
plt.show()
```

```python
sns.barplot(
    x=trope_counts.index,
    y=trope_counts.values
)

plt.title("Frequency of Narrative Tropes")
plt.xlabel("Trope")
plt.ylabel("Number of Books")

plt.show()
trope_counts = trope_counts.sort_values(ascending=False)
```

```python
df.groupby("magic")["rating"].mean().plot(kind="bar")
plt.title("Average Rating: Magic vs Non-Magic Books")
plt.show()
```

```python
trope_summary.plot(x="trope", y="avg_rating", kind="bar")
plt.title("Average Rating by Story Trope")
plt.show()
```

```python
sns.barplot(
    data=trope_summary,
    x="trope",
    y="avg_rating"
)

plt.title("Average Rating by Narrative Trope")
plt.xlabel("Trope")
plt.ylabel("Average Rating")

plt.show()
```

```python
plt.scatter(df["ratings"], df["rating"], alpha=0.5)

plt.xscale("log")

plt.xlabel("Number of Ratings (log scale)")
plt.ylabel("Average Rating")
plt.title("Book Popularity vs Reader Rating")

plt.show()
```

```python
df["primary_genre"] = df["genres"].str.extract(r"'(.*?)'")
```

```python
top_genres = df["primary_genre"].value_counts().head(10)

sns.barplot(
    x=top_genres.values,
    y=top_genres.index
)

plt.title("Top Genres in Dataset")
plt.xlabel("Number of Books")
plt.ylabel("Genre")

plt.show()
```

```python
corr = df[["rating","ratings","pages"]].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Between Key Variables")

plt.show()
```
